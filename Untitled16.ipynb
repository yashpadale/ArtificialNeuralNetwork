First defining a path to the dataset and creating id to text dictionary to maintain the information of which audio file belongs to which text as it has a metadata.csv which tell us abt the audio and the respective text which is required to teach the transformer.The function defined is used to return mapping of audio paths and transcription texts.The limit set for transcription text is 200. We can download the data in this colab and then train the model but the file takes long time to download and also takes google drive space .

```python
keras.utils.get_file(
    os.path.join(os.getcwd(), "data.tar.gz"),
    "https://data.keithito.com/data/speech/LJSpeech-1.1.tar.bz2",
    extract=True,
    archive_format="tar",
    cache_dir=".",
)


saveto = "./datasets/LJSpeech-1.1"
wavs = glob("{}/**/*.wav".format(saveto), recursive=True)

id_to_text = {}
with open(os.path.join(saveto, "metadata.csv"), encoding="utf-8") as f:
    for line in f:
        id = line.strip().split("|")[0]
        text = line.strip().split("|")[2]
        id_to_text[id] = text


def get_data(wavs, id_to_text, maxlen=50):
    """returns mapping of audio paths and transcription texts"""
    data = []
    for w in wavs:
        id = w.split("/")[-1].split(".")[0]
        if len(id_to_text[id]) < maxlen:
            data.append({"audio": w, "text": id_to_text[id]})
    return data
max_target_len = 200
data = get_data(wavs, id_to_text, max_target_len)
```

It is important to create the numbers into vectors. they are numerical representation of a number in layman language they are used to train the model as it can not do math on string data .The vocab size tells us abt the size of the vocabulary as the name suggests.

```python
class VectorizeChar:
    def __init__(self, max_len=50):
        self.vocab = (
            ["-", "#", "<", ">"]
            + [chr(i + 96) for i in range(1, 27)]
            + [" ", ".", ",", "?"]
        )
        self.max_len = max_len
        self.char_to_idx = {}
        for i, ch in enumerate(self.vocab):
            self.char_to_idx[ch] = i

    def __call__(self, text):
        text = text.lower()
        text = text[: self.max_len - 2]
        text = "<" + text + ">"
        pad_len = self.max_len - len(text)
        return [self.char_to_idx.get(ch, 1) for ch in text] + [0] * pad_len

    def get_vocabulary(self):
        return self.vocab
vectorizer = VectorizeChar(max_target_len)
print("vocab size", len(vectorizer.get_vocabulary()))
```

Functions to preprocess audio and text data, convert them into TensorFlow datasets, and create a TensorFlow dataset suitable for training a model.It reads and processes audio files into spectrograms, tokenizes text transcriptions, and creates TensorFlow datasets that are batched and prefetched for training.

```python

def create_text_ds(data):
    texts = [_["text"] for _ in data]
    text_ds = [vectorizer(t) for t in texts]
    text_ds = tf.data.Dataset.from_tensor_slices(text_ds)
    return text_ds


def path_to_audio(path):
    # spectrogram using stft
    audio = tf.io.read_file(path)
    audio, _ = tf.audio.decode_wav(audio, 1)
    audio = tf.squeeze(audio, axis=-1)
    stfts = tf.signal.stft(audio, frame_length=200, frame_step=80, fft_length=256)
    x = tf.math.pow(tf.abs(stfts), 0.5)
    # normalisation
    means = tf.math.reduce_mean(x, 1, keepdims=True)
    stddevs = tf.math.reduce_std(x, 1, keepdims=True)
    x = (x - means) / stddevs
    audio_len = tf.shape(x)[0]
    # padding to 10 seconds
    pad_len = 2754
    paddings = tf.constant([[0, pad_len], [0, 0]])
    x = tf.pad(x, paddings, "CONSTANT")[:pad_len, :]
    return x


def create_audio_ds(data):
    flist = [_["audio"] for _ in data]
    audio_ds = tf.data.Dataset.from_tensor_slices(flist)
    audio_ds = audio_ds.map(path_to_audio, num_parallel_calls=tf.data.AUTOTUNE)
    return audio_ds


def create_tf_dataset(data, bs=4):
    audio_ds = create_audio_ds(data)
    text_ds = create_text_ds(data)
    ds = tf.data.Dataset.zip((audio_ds, text_ds))
    ds = ds.map(lambda x, y: {"source": x, "target": y})
    ds = ds.batch(bs)
    ds = ds.prefetch(tf.data.AUTOTUNE)
    return ds
split = int(len(data) * 0.99)
train_data = data[:split]
test_data = data[split:]
ds = create_tf_dataset(train_data, bs=16)
val_ds = create_tf_dataset(test_data, bs=4)
# we have created the train test and val ds
```

Creating a Token embedding class for embedding token sequences,incorporating both token embeddings and positional embeddings. Provides both the content i.e. the token and positional i.e. the order information of sequences.

```python
class TokenEmbedding(layers.Layer):
    def __init__(self, num_vocab=1000, maxlen=100, num_hid=64):
        super().__init__()
        self.emb = keras.layers.Embedding(num_vocab, num_hid)
        self.pos_emb = layers.Embedding(input_dim=maxlen, output_dim=num_hid)

    def call(self, x):
        maxlen = tf.shape(x)[-1]
        x = self.emb(x)
        positions = tf.range(start=0, limit=maxlen, delta=1)
        positions = self.pos_emb(positions)
        return x + positions
```

This layer is for extracting hierarchical features from the input speech data. Each convolutional layer detects different patterns in the input sequence, and the series of convolutions allows the model to learn more complex features as the data passes through each layer.

```python

class SpeechFeatureEmbedding(layers.Layer):
    def __init__(self, num_hid=64, maxlen=100):
        super().__init__()
        self.conv1 = keras.layers.Conv1D(
            num_hid, 11, strides=2, padding="same", activation="relu"
        )
        self.conv2 = keras.layers.Conv1D(
            num_hid, 11, strides=2, padding="same", activation="relu"
        )
        self.conv3 = keras.layers.Conv1D(
            num_hid, 11, strides=2, padding="same", activation="relu"
        )

    def call(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        return self.conv3(x)
```

Process sequences of data by applying self-attention and feed-forward transformations. It captures the dependencies and relationships within the input sequence and transforms it into a richer representation.

```python

class TransformerEncoder(layers.Layer):
    def __init__(self, embed_dim, num_heads, feed_forward_dim, rate=0.1):
        super().__init__()
        self.att = layers.MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)
        self.ffn = keras.Sequential(
            [
                layers.Dense(feed_forward_dim, activation="relu"),
                layers.Dense(embed_dim),
            ]
        )
        self.layernorm1 = layers.LayerNormalization(epsilon=1e-6)
        self.layernorm2 = layers.LayerNormalization(epsilon=1e-6)
        self.dropout1 = layers.Dropout(rate)
        self.dropout2 = layers.Dropout(rate)

    def call(self, inputs, training=False):
        attn_output = self.att(inputs, inputs)
        attn_output = self.dropout1(attn_output, training=training)
        out1 = self.layernorm1(inputs + attn_output)
        ffn_output = self.ffn(out1)
        ffn_output = self.dropout2(ffn_output, training=training)
        return self.layernorm2(out1 + ffn_output)

```

Processes the target sequence in conjunction with the encoder output to generate the final decoded sequence. It combines self-attention to process the target sequence, encoder-decoder attention to incorporate information from the encoded input sequence, and a feed-forward network to produce the final output.

```python
class TransformerDecoder(layers.Layer):
    def __init__(self, embed_dim, num_heads, feed_forward_dim, dropout_rate=0.1):
        super().__init__()
        self.layernorm1 = layers.LayerNormalization(epsilon=1e-6)
        self.layernorm2 = layers.LayerNormalization(epsilon=1e-6)
        self.layernorm3 = layers.LayerNormalization(epsilon=1e-6)
        self.self_att = layers.MultiHeadAttention(
            num_heads=num_heads, key_dim=embed_dim
        )
        self.enc_att = layers.MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)
        self.self_dropout = layers.Dropout(0.5)
        self.enc_dropout = layers.Dropout(0.1)
        self.ffn_dropout = layers.Dropout(0.1)
        self.ffn = keras.Sequential(
            [
                layers.Dense(feed_forward_dim, activation="relu"),
                layers.Dense(embed_dim),
            ]
        )

    def causal_attention_mask(self, batch_size, n_dest, n_src, dtype):
        i = tf.range(n_dest)[:, None]
        j = tf.range(n_src)
        m = i >= j - n_src + n_dest
        mask = tf.cast(m, dtype)
        mask = tf.reshape(mask, [1, n_dest, n_src])
        mult = tf.concat(
            [tf.expand_dims(batch_size, -1), tf.constant([1, 1], dtype=tf.int32)], 0
        )
        return tf.tile(mask, mult)

    def call(self, enc_out, target):
        input_shape = tf.shape(target)
        batch_size = input_shape[0]
        seq_len = input_shape[1]
        causal_mask = self.causal_attention_mask(batch_size, seq_len, seq_len, tf.bool)
        target_att = self.self_att(target, target, attention_mask=causal_mask)
        target_norm = self.layernorm1(target + self.self_dropout(target_att))
        enc_out = self.enc_att(target_norm, enc_out)
        enc_out_norm = self.layernorm2(self.enc_dropout(enc_out) + target_norm)
        ffn_out = self.ffn(enc_out_norm)
        ffn_out_norm = self.layernorm3(enc_out_norm + self.ffn_dropout(ffn_out))
        return ffn_out_norm
```

This model is designed to process speech input, represented as spectrograms, and produce a sequence of output tokens, such as text. It implements both the encoder and decoder parts of the Transformer architectur

```python

class Transformer(keras.Model):
    def __init__(
        self,
            num_hid=64,
            num_head=2,
            num_feed_forward=128,
            source_maxlen=100,
            target_maxlen=100,
            num_layers_enc=4,
            num_layers_dec=1,
            num_classes=10,
    ):
        super().__init__()
        self.loss_metric = keras.metrics.Mean(name="loss")
        self.num_layers_enc = num_layers_enc
        self.num_layers_dec = num_layers_dec
        self.target_maxlen = target_maxlen
        self.num_classes = num_classes

        self.enc_input = SpeechFeatureEmbedding(num_hid=num_hid, maxlen=source_maxlen)
        self.dec_input = TokenEmbedding(
            num_vocab=num_classes, maxlen=target_maxlen, num_hid=num_hid
        )

        self.encoder = keras.Sequential(
            [self.enc_input]
            + [
                TransformerEncoder(num_hid, num_head, num_feed_forward)
                for _ in range(num_layers_enc)
            ]
        )

        for i in range(num_layers_dec):
            setattr(
                self,
                f"dec_layer_{i}",
                TransformerDecoder(num_hid, num_head, num_feed_forward),
            )

        self.classifier = layers.Dense(num_classes)

    def decode(self, enc_out, target):
        y = self.dec_input(target)
        for i in range(self.num_layers_dec):
            y = getattr(self, f"dec_layer_{i}")(enc_out, y)
        return y

    def call(self, inputs):
        source = inputs[0]
        target = inputs[1]
        x = self.encoder(source)
        y = self.decode(x, target)
        return self.classifier(y)

    @property
    def metrics(self):
        return [self.loss_metric]

    def train_step(self, batch):
        source = batch["source"]
        target = batch["target"]
        dec_input = target[:, :-1]
        dec_target = target[:, 1:]
        with tf.GradientTape() as tape:
            preds = self([source, dec_input])
            one_hot = tf.one_hot(dec_target, depth=self.num_classes)
            mask = tf.math.logical_not(tf.math.equal(dec_target, 0))
            loss = model.compute_loss(None, one_hot, preds, sample_weight=mask)
        trainable_vars = self.trainable_variables
        gradients = tape.gradient(loss, trainable_vars)
        self.optimizer.apply_gradients(zip(gradients, trainable_vars))
        self.loss_metric.update_state(loss)
        return {"loss": self.loss_metric.result()}

    def test_step(self, batch):
        source = batch["source"]
        target = batch["target"]
        dec_input = target[:, :-1]
        dec_target = target[:, 1:]
        preds = self([source, dec_input])
        one_hot = tf.one_hot(dec_target, depth=self.num_classes)
        mask = tf.math.logical_not(tf.math.equal(dec_target, 0))
        loss = model.compute_loss(None, one_hot, preds, sample_weight=mask)
        self.loss_metric.update_state(loss)
        return {"loss": self.loss_metric.result()}

    @tf.function(input_signature=[tf.TensorSpec(shape=[None, None, None], dtype=tf.float32),
                                  tf.TensorSpec(shape=[], dtype=tf.int32)])
    def generate(self, source, target_start_token_idx):
        """Performs inference over one batch of inputs using greedy decoding."""
        bs = tf.shape(source)[0]
        enc = self.encoder(source)

        dec_input = tf.ones((bs, 1), dtype=tf.int32) * target_start_token_idx
        dec_logits = []

        for i in range(self.target_maxlen - 1):
            dec_out = self.decode(enc, dec_input)
            logits = self.classifier(dec_out)
            logits = tf.argmax(logits, axis=-1, output_type=tf.int32)
            last_logit = tf.expand_dims(logits[:, -1], axis=-1)
            dec_logits.append(last_logit)
            dec_input = tf.concat([dec_input, last_logit], axis=-1)
        return dec_input
```

The DisplayOutputs callback helps to monitor the model's performance visually by printing out the actual and predicted sequences during training. This can provide insights into how well the model is learning to generate the target sequences over time.

```python

class DisplayOutputs(keras.callbacks.Callback):
    def __init__(
        self, batch, idx_to_token, target_start_token_idx=27, target_end_token_idx=28
    ):

        self.batch = batch
        self.target_start_token_idx = target_start_token_idx
        self.target_end_token_idx = target_end_token_idx
        self.idx_to_char = idx_to_token

    def on_epoch_end(self, epoch, logs=None):
        if epoch % 5 != 0:
            return
        source = self.batch["source"]
        target = self.batch["target"].numpy()
        bs = tf.shape(source)[0]
        preds = self.model.generate(source, self.target_start_token_idx)
        preds = preds.numpy()
        for i in range(bs):
            target_text = "".join([self.idx_to_char[_] for _ in target[i, :]])
            prediction = ""
            for idx in preds[i, :]:
                prediction += self.idx_to_char[idx]
                if idx == self.target_end_token_idx:
                    break
            print(f"target:     {target_text.replace('-','')}")
            print(f"prediction: {prediction}\n")
```

The CustomSchedule class defines a custom learning rate schedule for training a model using TensorFlow's tf.keras.optimizers.schedules.LearningRateSchedule. The purpose of this custom schedule is to control the learning rate dynamically during training, incorporating both a warmup phase and a decay phase.

```python
class CustomSchedule(tf.keras.optimizers.schedules.LearningRateSchedule):
    def __init__(
        self,
            init_lr=0.00001,
            lr_after_warmup=0.001,
            final_lr=0.00001,
            warmup_epochs=15,
            decay_epochs=85,
            steps_per_epoch=203,
    ):
        super().__init__()
        self.init_lr = init_lr
        self.lr_after_warmup = lr_after_warmup
        self.final_lr = final_lr
        self.warmup_epochs = warmup_epochs
        self.decay_epochs = decay_epochs
        self.steps_per_epoch = steps_per_epoch

    def calculate_lr(self, epoch):
        warmup_lr = (
            self.init_lr
            + ((self.lr_after_warmup - self.init_lr) / (self.warmup_epochs - 1)) * epoch
        )
        decay_lr = tf.math.maximum(
            self.final_lr,
            self.lr_after_warmup
            - (epoch - self.warmup_epochs)
            * (self.lr_after_warmup - self.final_lr)
            / self.decay_epochs,
        )
        return tf.math.minimum(warmup_lr, decay_lr)

    def __call__(self, step):
        epoch = step // self.steps_per_epoch
        epoch = tf.cast(epoch, "float32")
        return self.calculate_lr(epoch)

    def get_config(self):
        return {
            'init_lr': self.init_lr,
            'lr_after_warmup': self.lr_after_warmup,
            'final_lr': self.final_lr,
            'warmup_epochs': self.warmup_epochs,
            'decay_epochs': self.decay_epochs,
            'steps_per_epoch': self.steps_per_epoch
        }
```

```python

batch = next(iter(val_ds))

idx_to_char = vectorizer.get_vocabulary()
display_cb = DisplayOutputs(
    batch, idx_to_char, target_start_token_idx=2, target_end_token_idx=3
)
model = Transformer(
    num_hid=200,
    num_head=2,
    num_feed_forward=400,
    target_maxlen=max_target_len,
    num_layers_enc=4,
    num_layers_dec=1,
    num_classes=34,
)
loss_fn = keras.losses.CategoricalCrossentropy(
    from_logits=True,
    label_smoothing=0.1,
)

learning_rate = CustomSchedule(
    init_lr=0.00001,
    lr_after_warmup=0.001,
    final_lr=0.00001,
    warmup_epochs=15,
    decay_epochs=85,
    steps_per_epoch=len(ds),
)
optimizer = keras.optimizers.Adam(learning_rate)
model.compile(optimizer=optimizer, loss=loss_fn)


```

```python
from pytube import YouTube
from google.colab import drive
import os
from moviepy.editor import VideoFileClip
import gradio as gr
import numpy as np
import librosa

def detect_voice_activity(signal, sr, threshold, chunk_duration):
    energy = np.sum(np.square(signal))

    chunk_size = int(sr * chunk_duration)
    voice_segments = []

    for i in range(0, len(signal), chunk_size):
        # Calculate energy of current chunk
        chunk_energy = np.sum(np.square(signal[i:i+chunk_size]))

        # Calculate energy ratio
        energy_ratio = chunk_energy / energy

        # Check if energy ratio is above threshold
        if energy_ratio >= threshold:
            # Add chunk to voice activity segments
            start_idx = i
            end_idx = min(i + chunk_size, len(signal))
            voice_segments.append((start_idx, end_idx))

    return voice_segments

import soundfile as sf
def process_audio_chunk(chunk_path):
    audio = tf.io.read_file(chunk_path)
    audio, _ = tf.audio.decode_wav(audio, desired_channels=1)
    audio = tf.squeeze(audio, axis=-1)
    stfts = tf.signal.stft(audio, frame_length=200, frame_step=80, fft_length=256)
    x = tf.math.pow(tf.abs(stfts), 0.5)

    means = tf.math.reduce_mean(x, axis=1, keepdims=True)
    stddevs = tf.math.reduce_std(x, axis=1, keepdims=True)
    x = (x - means) / stddevs

    pad_len = 2754
    paddings = tf.constant([[0, pad_len], [0, 0]])
    x = tf.pad(x, paddings, "CONSTANT")[:pad_len, :]

    return x

def divide_wav_into_chunks(wav_file, output_dir, threshold=0.1, chunk_duration=10):
    signal, sr = librosa.load(wav_file, sr=None)
    voice_segments = detect_voice_activity(signal, sr, threshold, chunk_duration)

    chunk_info_list = []

    for i, (start_idx, end_idx) in enumerate(voice_segments):
        chunk_id = i + 1
        chunk_length = (end_idx - start_idx) / sr
        text = ""
        start_time = start_idx / sr
        end_time = end_idx / sr

        chunk_info = {
            "chunk_id": chunk_id,
            "chunk_length": chunk_length,
            "text": text,
            "start_time": start_time,
            "end_time": end_time
        }

        chunk_info_list.append(chunk_info)

        chunk = signal[start_idx:end_idx]
        chunk_path = os.path.join(output_dir, f"chunk_{i}.wav")
        sf.write(chunk_path, chunk, sr)

        print(f"Chunk {i} saved: {chunk_path}")

        # Process the audio chunk
        processed_chunk = process_audio_chunk(chunk_path)
        print(f"Processed chunk {i} shape: {processed_chunk.shape}")

    return chunk_info_list


wav="/content/drive/MyDrive/WAV_/LJSpeech DDC test.wav"
output_directory = "/content/drive/MyDrive/chunks"

lst=divide_wav_into_chunks(wav,output_directory)
print(lst)
def predict(output_directory,lst):
  pred=[]
  for i in range(len(lst)):
    d=lst[i]
    audio_path_1=d["chunk_id"]-1
    chunk = output_directory + "/chunk_" + str(audio_path_1) + ".wav"
    text="Random Stuff"
    pred.append({'audio':chunk,'text':text})
  return pred


pred=predict(output_directory,lst)

x = create_tf_dataset(pred)
def generate(x,lst):
  for element in x:
    source = element['source'].numpy()
    target = element['target'].numpy()

    batch_size = source.shape[0]
    for i in range(batch_size):
        single_source = tf.convert_to_tensor([source[i]])
        single_target = tf.convert_to_tensor([target[i]])
        target_start_token_idx = tf.convert_to_tensor([2], dtype=tf.int32)
        output = generate_signature(source=single_source, target_start_token_idx=target_start_token_idx)
        output_tensor = output['output_0']
        vocab = vectorizer.get_vocabulary()
        predicted_text = ""
        for idx in output_tensor.numpy()[0]:
            char = vocab[idx]
            if char == ">":
                break
            predicted_text += char
        print("Predicted text:", predicted_text)
        lst_=lst[i]
        lst_["text"]=predicted_text
  return lst

output=generate(x,lst)
print("output")
print(output)



```

The following code extracts the audio and text and saves them in a list from where i will use them as i know that each index has its corresponding audio and text pairs . This code may take long time to run as it will run till all the audio text pairs are not stored in the list .

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

def web_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--headless")
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-gpu')
    options.add_argument("--window-size=1920,1200")
    options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome(options=options)
    return driver

def get_text_and_audio(driver):
    # Extract text from the content-container
    text_elements = driver.find_elements(By.CSS_SELECTOR, ".content-container")
    text = "\n".join([element.text for element in text_elements])

    # Extract audio links from the audio elements
    audio_elements = driver.find_elements(By.CSS_SELECTOR, ".audio-player")
    audio_links = [element.get_attribute("src") for element in audio_elements]

    return text, audio_links

driver = web_driver()
text_list=[]
audio_list=[]
try:
    driver.get("https://live.bible.is/bible/MARDPI/MAT/1")
    time.sleep(5)  # Add a delay to ensure the page is loaded

    # Loop through each page
    while True:
        # Get text and audio for the current page
        text, audio_links = get_text_and_audio(driver)
        text_list.append(text)
        audio_list.append(audio_links[0])
        # Click on the "Next" button using a CSS selector for the SVG arrow
        try:
            next_button = driver.find_element(By.CSS_SELECTOR, ".next-arrow-svg")
            next_button.click()
            time.sleep(5)  # Add a delay to ensure the next page is loaded
        except Exception as e:
            print("Next button not found or not clickable:", e)
            break
finally:
    driver.quit()
print(text_list)
print(audio_list)
print(len(text_list))
print(len(audio_list))
```

```python
import os
import librosa
import librosa.display
import matplotlib.pyplot as plt
import pandas as pd
from wordcloud import WordCloud
import seaborn as sns
import numpy as np
```

```python
audio_dir = "/content/drive/MyDrive/audio_files"
transcripts_dir = "/content/drive/MyDrive/text_files"
audio_files = [f for f in os.listdir(audio_dir) if f.endswith('.mp3')]
transcript_files = [f for f in os.listdir(transcripts_dir) if f.endswith('.txt')]

# Ensure the files are paired correctly
audio_files.sort()
transcript_files.sort()

# Function to load transcripts
def load_transcript(file_path):
    with open(file_path, 'r') as file:
        return file.read()

# Create a DataFrame to store the paths and transcripts
data = pd.DataFrame({
    'audio_path': [os.path.join(audio_dir, f) for f in audio_files],
    'transcript': [load_transcript(os.path.join(transcripts_dir, f)) for f in transcript_files]
})
```

```python

def extract_audio_features(file_path):
    y, sr = librosa.load(file_path, sr=None)
    duration = librosa.get_duration(y=y, sr=sr)
    return {'duration': duration, 'sampling_rate': sr}


audio_features = data['audio_path'].apply(extract_audio_features)
audio_features_df = pd.DataFrame(audio_features.tolist())


data = pd.concat([data, audio_features_df], axis=1)


plt.figure(figsize=(10, 6))
sns.histplot(data['duration'], bins=30, kde=True)
plt.title('Distribution of Audio Durations')
plt.xlabel('Duration (seconds)')
plt.ylabel('Frequency')
plt.show()

sample_audio_path = data['audio_path'][0]   #  waveform and spectrogram visualizations  of the first audio only
y, sr = librosa.load(sample_audio_path, sr=None)

plt.figure(figsize=(14, 5))
librosa.display.waveshow(y, sr=sr)
plt.title('Waveform')
plt.show()

D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
plt.figure(figsize=(14, 5))
librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
plt.show()
```

```python

data['word_count'] = data['transcript'].apply(lambda x: len(x.split()))
data['char_count'] = data['transcript'].apply(lambda x: len(x))


plt.figure(figsize=(10, 6))
sns.histplot(data['word_count'], bins=30, kde=True)
plt.title('Distribution of Word Counts in Transcripts')
plt.xlabel('Word Count')
plt.ylabel('Frequency')
plt.show()


all_text = ' '.join(data['transcript'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Transcripts')
plt.show()
```
